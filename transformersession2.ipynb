```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
print("\n Step 3 : transformer model adn tokenizer loaded")
```

```python
input_text = "movie is not good"
input_tokens = tokenizer(input_text, return_tensors="pt")
```

```python
print("/nStep 4 :tokenization")
print('original text',input_text)
print('tokenized iputids',input_tokens['input_ids'])
print('tokenized attention mask',input_tokens['attention_mask'])
```

```python
#Embeddings
embeddings = model.model.encoder.embed_tokens(input_tokens['input_ids'])
print("/nStep 5 : embeddings")
print(embeddings.shape)
print(embeddings[0][0].tolist())
```

```python
with torch.no_grad():
    encoder_outputs = model.get_encoder()(input_tokens['input_ids'])
```

```python
print('first token contetualized output',encoder_outputs.last_hidden_state[0][0].tolist())
```

```python
decoder_input_ids = torch.tensor([[tokenizer.eos_token_id]])
print(decoder_input_ids)
```

```python
with torch.no_grad():
  decoder_outputs = model.model.decoder.embed_tokens(decoder_input_ids)
```

```python
print(decoder_outputs.shape)
print(decoder_outputs[0][0].tolist())
```

```python
with torch.no_grad():
  cross_attention = encoder_outputs.last_hidden_state @ decoder_outputs.transpose(1,2)
```

```python
print(cross_attention.shape)
print(cross_attention[0][0].tolist())
```

```python
output_tokens = model.generate(**input_tokens,max_length=20)
print('Generated tokens',output_tokens)
```

```python
translated_text = tokenizer.decode(output_tokens[0],skip_special_tokens=True)
print('Translated text  :  ',translated_text)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import DistilBertTokenizer,DiffLlamaForSequenceClassification,Trainer,TrainingArguments
from datasets import Dataset
```

```python
data = pd.read_csv('spam.csv',encoding='latin-1')
data = data[['label','text']]
data.columns = ['label','text']
```

```python
data['label'] = data['label'].map({'ham':0,'spam':1})
```

```python
train_df,val_df = train_test_split(data,test_size=0.2,random_state=42)
```

```python
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
```

```python
model_name = 'distilbert-base-uncased'
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DiffLlamaForSequenceClassification.from_pretrained(model_name,num_labels=2)
```

```python
def tokenize_function(examples):
  return tokenizer(examples['text'],padding='max_length',truncation=True)
train_dataset = train_dataset.map(tokenize_function,batched=True)
val_dataset = val_dataset.map(tokenize_function,batched=True)
```

```python
train_dataset.set_format(type='torch',columns=['input_ids','attention_mask','label'])
val_dataset.set_format(type='torch',columns=['input_ids','attention_mask','label'])
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,

)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)
```

```python
trainer.train()
```

```python
trainer.evealuate()
```

```python
model.save_pretrained('./spam_classifier')
tokenizer.save_pretrained('./spam_classifier')
```

```python
model_path = './spam_classifier'
tokenizer = DistilBertTokenizer.from_pretrained(model_path)
model = DiffLlamaForSequenceClassification.from_pretrained(model_path)
```

```python
def predict_text(text):
  inputs = tokenizer(text,padding=True,truncation=True,return_tensors='pt')
  model.eval()
  with torch.no_grad():
    outputs = model(**inputs)
  predicted_label = torch.argmax(outputs.logits,dim=1).item()
  if predicted_label == 0:
    predicted_class = 'ham'
  else:
    predicted_class = 'spam'
  return predicted_class
```

```python
predict_text("congratulations you won a lottery")
```
